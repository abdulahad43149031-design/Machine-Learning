```python
import pandas as pd
from sklearn.datasets import load_iris
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
df.shape
```

```python
df['species'].value_counts()
```

```python
df = df[df['species'] != 'virginica']
```

```python
df['species']  = df['species'].map({
    'setosa': 0,
    'versicolor': 1
})
```

```python
df.head(3)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X = df.drop('species', axis = 1)
y = df['species']
```

```python
X_train_full, X_test, y_train_full, y_test = train_test_split(X, y, test_size= 0.3, random_state = 42)
X_train, X_val, y_train, y_val = train_test_split(X_train_full, y_train_full, test_size = 0.3, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

models = {}
solvers = ['lbfgs', 'liblinear', 'newton-cg']

for solver in solvers:
    model = LogisticRegression(solver = solver)
    model.fit(X_train, y_train)
    
    y_val_pred = model.predict(X_val)
    score = accuracy_score(y_val, y_val_pred)
    
    models[solver] = score   # use hyperparameter value as key

models
```

```python
from sklearn.metrics import classification_report
model_best = LogisticRegression(solver='lbfgs', max_iter = 30)
model_best.fit(X_train_full, y_train_full)
y_pred = model_best.predict(X_test)
accuracy_score(y_test, y_pred)
print(classification_report(y_test, y_pred))
```
